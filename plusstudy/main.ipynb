### 단축평가 연습하기
> 단축평가는 논리연산에서 결과가 확실할 때 나머지 연산(평가)를 건너뛰는 기능
> and 그리고 or 

```python
result = False and print('이 뒤에 문장은 출력되지 않습니다!')
result = True and print('이 뒤에 문장은 출력됩니다!')
print(result)
```

```python
result = False or print('이 뒤에 문장은 출력됩니다!')
result = True or print('이 뒤에 문장은 출력되지 않습니다!')
print(result)
```

```python
my_list = [1, 2, 3]
# 위 리스트에 요소들이 존재할 때, 첫번째 요소를 반환받기.
            # 참으로 평가 and  
first_item = my_list and my_list[0]
print(first_item)
```
