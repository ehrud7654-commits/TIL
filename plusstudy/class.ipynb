```python
class AndroidSmartphone:
    os = 'Android'
    version = 1.2
    def __init__(self, model, price):
        print('생성자 호출됨')
        # 나만의 객체 정보
        self.model = model
        self.price = price

    # 클래스가 가질 수 있는 동작을 정의할 수 있다.
    # 동작 -> 함수 -> 클래스 내부에 있으니까 메서드라고 부른다.

    def show_info(self):
        return f'{self.model} & {self.price}'
    
    # 일정 버전 이하면, 전체 기종에게 업데이트를 강제로 적용한다.
    @classmethod
    def change_version(cls, ver):
        # 특정 version보다 낮은 경우에만(if)
        cls.version = ver
    
    # 스태틱 메서드를 만든다면?

    @staticmethod
    def is_expensive(price): #
        return price > 1000


my_phone = AndroidSmartphone('galaxys25', 1496000) #instance (생성자)

my_phone.show_info()
AndroidSmartphone.is_expensive(1496000)
print(my_phone.version)
AndroidSmartphone.change_version(2.2)
print(my_phone.version)



# 아이폰 -> 운영체제
# 안드로이드
# 인스턴스들 마다 각자 가지는, 변수들 -> 모델명, 가격, 출시일
```

```python
#MRO 확인법
#다중상속은 신중하게 한다

class Top:
    def __init__(self,name):
        self.name = name
    def method(self):
        print('Top')

class Left(Top):
    def method(self):
        print('Left')
        super().method()

class Right(Top):
    def method(self):
        print('Right')
        super().method()

class Bottom(Left, Right):
    def method(self):
        print('Bottom')
        super().method()


print(Bottom.mro())
b = Bottom(name = '이름')
b.method()
print(b.name)
```
