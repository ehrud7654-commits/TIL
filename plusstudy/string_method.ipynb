### 문자열 메소드 활용

- 문자열의 불변성 확인

- 파이썬의 문자열 메소드는 원본을 수정하지 않고, 새로은 객체를 생성합니다.
- 각 메소드는 실행 후 `id()` 값을 확인하여 메모리 주소가 달라지는 것을 확인해보자.

```python
s = '    python programming    '
print(f'오리지날 ID: {id(s)}')
s_strip = s.strip()
print(f'{s_strip} ID: {id(s_strip)}')
s_replace = s_strip.replace('python', 'js')
print(f'{s_replace} ID: {id(s_replace)}')
s_list = s_replace.split()
print(s_list)
s_join = '-'.join(s_list)
print(s_join)


```
