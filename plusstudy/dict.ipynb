# 딕셔너리
- 키와 값을 쌍으로 갖는다.
- 파이썬에서 딕셔너리를 정말 많이 씁니다.
- 삽입순서 정도는 보장해준다.(python 3.7 버전 이상)
- 가변형
- 빠른 탐색 보장 O(1)

```python
my_dict1 = {} 
my_dict2 = dict() # 생성자 

# 초기값이 있는 생성
student = {'name': 'isaac', 'age': 22, 'grade': 'A'}
```

```python
# 안전한 조회 `get(key[, default])`
# 키가 없을 때 에러(keyError)를 발생시키지 않는다. -> 기본값을 반환한다. 

scores = {'Lee':90, 'Shin': 100}
print(scores.get('Park'))
print(scores.get('Park', '없어요'))
```

```python
# 딕서너리 요소 확인하기(View)
# - 실시간 적용이 되기 때문
# keys(). items(), values()
# 딕서너리의 구성요소들을 실시간으로 반영해서 보여줍니다.
scores = {'Lee':90, 'Shin': 100}
k = scores.keys()
v = scores.values()
i = scores.items()
print(k)
print(type(k))
print(len(k))
for i in k:
    print(i)
```

```python
# 수정 `Update`
# 다른 딕셔너리들끼리 병합하거나, 한번에 업데이트할 때 사용할 수 있다.
scores1 = {'Lee':90, 'Shin': 100}
scores1.update({'Park':90, 'Shin': 80})
scores1
```

# 실전 활용 패턴 

```python
student = {'name': 'isaac', 'age': 22, 'grade': 'A'}

for key, value in student.items():
    print(f'{key}: {value}')
```

```python
prices = {'apple': 1000, 'banana': 1500}
for fruit, price in prices.items():
    print(f'{fruit}: {price}')
```

```python
prices = {'apple': 1000, 'banana': 1500}
expensive = {} # 값이 1200원을 넘는 과일 모으기 

for k, v in prices.items():
    if v >= 1200:
        expensive[k] = v

# print(expensive)

expensive = {k:v for k, v in prices.items() if v >= 1200}
expensive
```

```python
# setdefault, defaultdict
# 1. setdefault

text = ['kiwi', 'kiwi', 'kiwi']
counter = {}

for item in text:
    counter.setdefault(item, 0) # 키가 없으면, 0으로 초기화 후 조회
    counter[item] +=1


from collections import defaultdict
counter_dd = defaultdict(int) # 기본값을 0(int의 기본값)으로 자동 설정
for item in text:
    counter_dd[item] += 1
```

```python
# defaultdict 응용
from collections import defaultdict

counter_dd = defaultdict(lambda: 1) # 기본값을 0(int의 기본값)으로 자동 설정
for item in text:
    counter_dd[item] += 1




```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
target = 5
found = False # flag 변수

for r in range(len(matrix)):
    for c in range(len(matrix[r])):
        if matrix[r][c] == target:
            print(f"찾았다! 위치: {r}행 {c}열")
            found = True
            break # 내측 루프 탈출
    if found:
        break # 외측 루프 탈출
```
