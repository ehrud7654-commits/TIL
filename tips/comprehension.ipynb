```python
# [:] 슬라이싱을 통해서 얕은 복사를 할 수 있다.
import copy

base =[[1, 2], [3, 4]]
slice_copy = base[:]
slice_copy2 = copy.copy(base)

print(id(base))
print(id(slice_copy))
print(id(base[0]) == id(slice_copy[0]) )
```

```python
deep_copy = copy.deepcopy(base) # 깊은 복사
```

```python
# 수동으로 깊은복사 
manual = [item[:] for item in base]
print(id(base[0]) == id(manual[0]))
```
